```python
import pandas as pd
import matplotlib.pyplot as plt   # ✅ This line was missing
import seaborn as sns
from google.colab import files

# Step 1: Upload and load dataset
uploaded = files.upload()   # choose Tweets.csv
df = pd.read_csv("Tweets.csv")

# Step 2: Show first few rows
df.head(10)

# ---------- PLOT 1 ----------
plt.figure(figsize=(6,4))
sns.countplot(x='airline_sentiment', data=df, palette='coolwarm')
plt.title('Overall Sentiment Distribution')
plt.xlabel('Sentiment Type')
plt.ylabel('Number of Tweets')
plt.show()

print("🧠 Insight: Most tweets are negative, showing that passengers expressed more dissatisfaction than satisfaction with airlines.")

# ---------- PLOT 2 ----------
plt.figure(figsize=(8,4))
sns.countplot(x='airline', hue='airline_sentiment', data=df, palette='Set2')
plt.title('Sentiment Distribution by Airline')
plt.xlabel('Airline')
plt.ylabel('Tweet Count')
plt.legend(title='Sentiment', loc='upper right')
plt.show()

print("🧠 Insight: Some airlines (like United or US Airways) received more negative tweets, while others have a balanced sentiment spread.")

# ---------- PLOT 3 ----------
plt.figure(figsize=(7,4))
sns.countplot(y='negativereason', data=df, order=df['negativereason'].value_counts().index, palette='mako')
plt.title('Top Negative Reasons Mentioned by Users')
plt.xlabel('Count')
plt.ylabel('Negative Reason')
plt.show()

print("🧠 Insight: The most common complaints are about flight delays and customer service issues, indicating major areas for improvement.")

# ---------- PLOT 4 ----------
plt.figure(figsize=(6,4))
sns.histplot(df['airline_sentiment_confidence'], bins=20, kde=True, color='purple')
plt.title('Distribution of Sentiment Confidence')
plt.xlabel('Confidence Score')
plt.ylabel('Frequency')
plt.show()

print("🧠 Insight: Most sentiment predictions have high confidence (above 0.8), meaning the data labels are reliable for model training.")

# ---------- PLOT 5 ----------
plt.figure(figsize=(8,4))
sns.countplot(y='user_timezone', data=df, order=df['user_timezone'].value_counts().head(10).index, palette='Spectral')
plt.title('Top 10 User Timezones of Tweet Authors')
plt.xlabel('Count')
plt.ylabel('User Timezone')
plt.show()

print("🧠 Insight: Most tweets come from US timezones, showing that the dataset primarily represents American airline customers.")

```
