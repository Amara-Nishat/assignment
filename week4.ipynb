```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Load cleaned dataset
df = pd.read_csv("Tweets.csv")

# Map target to numeric
sentiment_map = {'negative': 0, 'neutral': 1, 'positive': 2}
df['sentiment_numeric'] = df['airline_sentiment'].map(sentiment_map)

# Select numeric columns
numeric_cols = ['airline_sentiment_confidence', 'negativereason_confidence', 'retweet_count', 'sentiment_numeric']
df_numeric = df[numeric_cols]

# ---------- Step 1: Compute correlation matrix ----------
corr = df_numeric.corr()

# ---------- Step 2: Show correlation with target ----------
corr_target = corr['sentiment_numeric'].sort_values(ascending=False)
print("Correlation with target (sentiment_numeric):\n", corr_target)

# ---------- Step 3: Heatmap ----------
plt.figure(figsize=(6,4))
sns.heatmap(corr, annot=True, cmap='coolwarm')
plt.title('Correlation Heatmap')
plt.show()

# ---------- Step 4: Top 3 features ----------
top_features = corr_target.drop('sentiment_numeric').head(3)
print("\n✅ Top 3 features most related to target variable:")
print(top_features)

```
