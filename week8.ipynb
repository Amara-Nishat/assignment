```python
# Import required libraries
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA

# Load dataset
df = pd.read_csv("Tweets.csv")
df = df.dropna(subset=['text'])  # Remove missing text values

# Step 1: Convert text to numerical features using TF-IDF
vectorizer = TfidfVectorizer(stop_words='english', max_features=1000)
X = vectorizer.fit_transform(df['text'])

# Step 2: Apply KMeans clustering
kmeans = KMeans(n_clusters=3, random_state=42)
kmeans.fit(X)

# Step 3: Reduce dimensions for visualization (PCA → 2D)
pca = PCA(n_components=2, random_state=42)
reduced_data = pca.fit_transform(X.toarray())

# Step 4: Plot the clusters
plt.figure(figsize=(10,6))
plt.scatter(reduced_data[:,0], reduced_data[:,1], c=kmeans.labels_, cmap='rainbow', alpha=0.7)
plt.title("K-Means Clustering of Tweets (2D PCA Visualization)")
plt.xlabel("Principal Component 1")
plt.ylabel("Principal Component 2")
plt.show()
```
