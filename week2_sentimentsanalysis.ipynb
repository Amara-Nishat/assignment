```python
import pandas as pd
from google.colab import files

# Upload your dataset file (e.g., Tweets.csv)
uploaded = files.upload()

# Read the uploaded file
df = pd.read_csv('Tweets.csv')   # change name if needed

# Show first 10 rows
df.head(10)
# Step 5: BEFORE CLEANING summary
print("Shape before cleaning:", df.shape)
print("\nMissing values before cleaning:\n", df.isnull().sum())
print("\nDuplicate rows before cleaning:", df.duplicated().sum())
print("\nBasic stats:\n", df.describe(include='all'))

# Step 6: Remove duplicate rows (if any)
df = df.drop_duplicates()

# Handle missing values

# Fill missing text columns with "Unknown"
df['negativereason'].fillna('Unknown', inplace=True)
df['tweet_location'].fillna('Unknown', inplace=True)
df['user_timezone'].fillna('Unknown', inplace=True)

# Drop columns that have too many missing values or are not useful
cols_to_drop = ['airline_sentiment_gold', 'negativereason_gold', 'tweet_coord']
df.drop(columns=cols_to_drop, inplace=True, errors='ignore')

# Step 7: AFTER CLEANING summary
print("\n✅ AFTER CLEANING SUMMARY")
print("Shape after cleaning:", df.shape)
print("\nMissing values after cleaning:\n", df.isnull().sum())
print("\nDuplicate rows after cleaning:", df.duplicated().sum())
print("\nBasic stats after cleaning:\n", df.describe(include='all'))


```
